```python
# Palindrome script
import math

def checkPallindrome(inputString: str) -> bool:
    """
    A function that determines if a string is pallindromic
    """

    # Get the length of the input string.
    inputLength = len(inputString)

    # Ensure that the user has provided a non-empty string.
    if inputLength == 0:
        raise ValueError("inputString must not be empty!")

    # Start the current index count at zero
    currentIndex = 0

    # Loop over half of the letters, you only need to goto the middle of the word. This will make a bigger difference with longer strings.
    # Floor is used to account for the odd numbered words. 
    while(currentIndex < math.floor(inputLength / 2)):
        # Check if character at front of the word is not the same as the one at the symetrical position. 
        if inputString[currentIndex] != inputString[-(currentIndex + 1)]:
            # The function can stop if a value is found to be non-symetical. No point wasting time...
            return False

        # Increment the count
        currentIndex += 1

    # If the script finishes the loop, then all values are symetrical and the word is pallindromic.
    return True

# Tests 
if checkPallindrome("ADDDA") == True:
    print("Test 1 (ADDDA): Pass!")

if checkPallindrome("ADDA") == True:
    print("Test 2 (ADDA): Pass!")

if checkPallindrome("ADDDasdA") == False:
    print("Test 3 (ADDDasdA): Pass!")

if checkPallindrome("qwertyuiopoiuytrewq") == True:
    print("Test 4 (qwertyuiopoiuytrewq): Pass!")
```

2. Write python code that is testing a random generator. Run the test for 20,000 samples.
Explain why the test results provides evidence that the random number generator is working
correctly.

```python
import random
import matplotlib.pyplot as plt

# A function that generates a list of random numbers between 0 and 1 of length n
def generateRandomNumbers(n):

    # Create a list to store all of the random numbers
    randomNumbers = list()

    # Generate a random number and append it to the list n times.
    for i in range(n):
        randomNumbers.append(random.random())
        
    # Finally, return the list of random numbers.
    return randomNumbers
        
# Create a list of 20,000 random numbers
listOfRands = generateRandomNumbers(20000)

# Visualize the distribution of the randoms
plt.hist(listOfRands, color = 'blue', edgecolor = 'black',
         bins = int(180/5))

```

It is clear to see from the plot that for the values in our list, we have a uniform distribution of values. However, we also need to see whether order in which they are generated, is also random, this can be investigated by plotting the index value of the random number against the value of the number.

```python
import pandas as pd

df = pd.DataFrame(data = {'y': listOfRands, 'x': range(0, len(listOfRands))})

ax = df.plot.scatter(x = 'x', y = 'y', figsize = (15, 6))
ax.set_xlabel("Index")
ax.set_ylabel("Value")
```

We see that the order in which the numbers are generated does not effect the likelihood of what the number is, as such, they are effectively random.
