6. How does 'for' loop and 'while' loop differ in Python and when do you choose to use them?

for loop - This type of loop is used to iterate over a collection or across a range. For loops will run the code in the enclosed scope once per item in the collection. The for loop will exit prematurely if a break statement is reached. For loops are best used when each item in a collection need to have the same logic applied to them.

while loop - The while loop will loop as long as the condition is true or until a break statement is reached. While loops are often used to repeat a process until a specific condition is met, for example, this may be used whilst asking for user input, whereby, the program loops until the user enters an appropriate response, at which point a break statement is used to continue with the program. 

7. How to reverse a list?

```python
# Python includes a built-in function which can reverse lists.
myList = [1, 2, 3, 4, 5, 6, 7, 8]
print(f"Forwards: {myList}")

myList.reverse()
print(f"Reversed: {myList}")

# Lists can also be reverse using slicing
myList = [1, 2, 3, 4, 5, 6, 7, 8]
print(f"Forwards: {myList}")
print(f"Reversed: {myList[::-1]}")

# I prefer the use of reverse slicing since calling the reverse function mutates
# the list which may lead to subtle errors which are hard to find.
```

1. Write a Python script that checks if genes on the Y or the X have on average more exons (Hint: Use
two functions â€“ one function getting the exon count, one function to calculate the average).

```python
#!/bin/python3
from typing import List

def getExonLengths(chromosomeID: str) -> List[int]:
    
    filePath = "./human-genes.gtf"

    with open(filePath, 'r') as geneFile:

        exonLengths = list()

        for line in geneFile:
            tmp = line.split("\t")

            if tmp[0].lower() == chromosomeID.lower():
                # This strips out the unwanted characters from the string.
                # It may be possible with regex, but that is a headache...
                exonLengths.append(int(tmp[8].split(";")[2][13:-1]))

        return exonLengths

def mean(inList: List[int]) -> float:
    sum = 0
    for val in inList:
        sum += val

    return float(sum) / float(len(inList))

yMeanExonLength = mean(getExonLengths('y'))
xMeanExonLength = mean(getExonLengths('x'))

print(f"Mean exon counts\nX: {xMeanExonLength}\nY: {yMeanExonLength}")
```

2. Write a Python script that tests the hypothesis of the scientist that the number of genes is relevant
for the severity of variation in chromosome numbers (Hints: Use dictionary to count genes and sort).
Summarize your finding in 2-3 sentences

```python
#!/usr/bin/python3
# Task: count genes per chromosome
#       print for every chromosome the number of genes starting with the
#       chromosome having the fewest genes and ending with the chromosome having
#       the most genes.

def getGeneCount():

    # Declare the location of the human-genes.gtf file
    filePath = "./human-genes.gtf"

    # Open the file with error handling
    with open(filePath, 'r') as geneFile:

        # Create an empty dictionary to store the chromosome names
        # and number of genes.
        genes = {}

        # Read the geneFile line by line
        for line in geneFile:

            # Split each line by tabs.
            tmp = line.split("\t")
            
            # Check if the key for the given chromosome exists in the dictionary
            if str(tmp[0]) in genes:
                # If the key exists, then increment the count by 1. This works
                # because each line in the file is a single gene.
                genes[str(tmp[0])] += 1
            else:
                # If the key does not exist, create one and set the count to 1.
                genes[str(tmp[0])] = 1

        # Return the dictionary of gene counts to the caller.
        return genes


# Use dictionary comprehension to sort the output of the getGeneCount function.
geneCount = {chromosome: count for chromosome, count in sorted(getGeneCount().items(), key = lambda item: item[1])}

def printGeneCount(countDict):
    for key in countDict:
        print(f"Chromosome {key} has {countDict[key]} genes")

printGeneCount(geneCount)
```
