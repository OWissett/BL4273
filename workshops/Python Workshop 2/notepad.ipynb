```python
# Finds the longest and the shortest gene in the human genome

filePath = "./human-genes.gtf"

with open(filePath, 'r') as geneFile:

    smallestLine = ''
    smallestSize = None

    longestLine = ''
    longestSize = None

    for line in geneFile:
        tmp = line.split("\t")
        start = int(tmp[3])
        end = int(tmp[4])

        length = end - start

        if (smallestSize is None) or (smallestSize > length):
            smallestSize = length
            smallestLine = line

        if (longestSize is None) or (longestSize < length):
            longestSize = length
            longestLine = line

        #print(length)

    print(f"Shortest gene length: {smallestSize}\nLongest gene size: {longestSize}")
```

```python
# How many genes X and Y have


filePath = "./human-genes.gtf"

with open(filePath, 'r') as geneFile:

    xCount = 0
    yCount = 0

    for line in geneFile:
        tmp = line.split("\t")

        if tmp[0].lower() == 'x':
            xCount += 1

        if tmp[0].lower() == 'y':
            yCount += 1

    print(f"The X chromosome has {xCount} genes, and the Y chromosome has {yCount} genes.")
```

```python
#!/bin/python3
from typing import List

def getExonLengths(chromosomeID: str) -> List[int]:
    
    filePath = "./human-genes.gtf"

    with open(filePath, 'r') as geneFile:

        exonLengths = list()

        for line in geneFile:
            tmp = line.split("\t")

            if tmp[0].lower() == chromosomeID.lower():
                # This strips out the unwanted characters from the string.
                # It may be possible with regex, but that is a headache...
                exonLengths.append(int(tmp[8].split(";")[2][13:-1]))

        return exonLengths

def mean(inList: List[int]) -> float:
    sum = 0
    for val in inList:
        sum += val

    return float(sum) / float(len(inList))

yMeanExonLength = mean(getExonLengths('y'))
xMeanExonLength = mean(getExonLengths('x'))

print(f"Mean exon counts\nX: {xMeanExonLength}\nY: {yMeanExonLength}")
```
